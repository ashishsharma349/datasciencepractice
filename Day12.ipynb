```python
import pandas as pd
import numpy as np
df=pd.read_excel(r"..\datasets\flight_price.xlsx")
```

```python
df.head()
```

```python
df['year']=df['Date_of_Journey'].str.split("/").str[0]
df['month']=df['Date_of_Journey'].str.split("/").str[1]
df['date']=df['Date_of_Journey'].str.split("/").str[2]
```

```python
df.head()
```

```python
df['date']=df['date'].astype(int)
df['month']=df['month'].astype(int)
df['year']=df['year'].astype(int)
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df["Arrival_Time"]=df["Arrival_Time"].apply(lambda x:x.split(" ")[0])
```

```python
df.head(5)
```

```python
df['arrival_hour']=df['Arrival_Time'].str.split(":").str[0]
df['arrival_minute']=df['Arrival_Time'].str.split(":").str[1]
df['arrival_hour']=df['arrival_hour'].astype(int)
df['arrival_minute']=df['arrival_minute'].astype(int)
```

```python
df['departure_hour']=df['Dep_Time'].str.split(":").str[0]
df['departure_minute']=df['Dep_Time'].str.split(":").str[1]
df['departure_hour']=df['departure_hour'].astype(int)
df['departure_minute']=df['departure_minute'].astype(int)
```

```python
df.drop(['Dep_Time','Arrival_Time'],axis=1,inplace=True)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'2 stops':2,'1 stop': 1, '3 stops': 3, np.nan: 1, '4 stops':4 })
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df['duration_hour'] = df['Duration'].str.split(" ").str[0].str.split("h").str[0].str.split("m").str[0]
df['duration_minute'] = df['Duration'].str.split(" ").str[1].str.split("m").str[0]
```

```python
# df["duration_minute"].fillna(df['duration_minute'].mean(),inplace=True)

df['duration_minute'].fillna(0, inplace=True)
df["duration_minute"].isna().sum()
```

```python
df.head(40)
```

```python
df['duration_hour'] =df['duration_hour'].astype(int)
df['duration_minute']=df['duration_minute'].astype(int)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
df_encoded=df_ohe=pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
df.drop(columns=['Airline','Source','Destination','Duration','Additional_Info'],axis=1,inplace=True)
```

```python
df_final=pd.concat([df,df_encoded],axis=1)
```
