## üîé **Quick Revision ‚Äì Decision Tree Basics**

* A decision tree splits data by asking **questions** (like ‚ÄúIs age > 30?‚Äù).
* Each split reduces **impurity** (Gini/Entropy).
* It continues splitting until it reaches **pure leaves** (or a stopping rule).
* Good for **classification** (Yes/No, categories) and **regression** (predicting numbers).

```python
from sklearn.datasets import load_iris
data=load_iris()
data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data['data'],data['target'],test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score
model=DecisionTreeClassifier(max_depth=2)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

### üëâ In practice:

If you want overall balanced performance ‚Üí use 'macro'.

If dataset is imbalanced ‚Üí use 'weighted'.

If you want per-class metrics ‚Üí set average=None.

```python
# y_pred
# confusion_matrix(y_test,y_pred)
print(precision_score(y_test,y_pred,average="macro"))# macro for multiclass classification
print(recall_score(y_test,y_pred,average="macro"))
print(f1_score(y_test,y_pred,average="macro"))
```
