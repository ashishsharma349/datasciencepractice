✅ Exercise: Polynomial Regression & Overfitting
Goal: Understand underfitting vs overfitting through polynomial regression.
Steps:
Generate synthetic data: y = sin(x) + noise
Fit three models:
Linear regression (1st degree)
Polynomial regression (3rd degree)
Polynomial regression (15th degree)
Plot all three and compare performance on training vs test data.

Ques-Bias and Variance trade-off-----------------------------------------------------------------------------
Ans- Bias is the inablity of ml model to truly capture the relationship in the data , high bias leads to overfitting and very low/low bias means underfitting , whereas when mode's performance vary greatly on train and test data it has high variance else low.
High variance means model is overfitting and very low/low variance signifies underfitting.

```python
import math
import random
import numpy as np
def gen_data(x):
    y=math.cos(x)
    return y
x=np.array([i for i in range(1,100)])
y=[]
for i in x:
    y.append(gen_data(i))
y=np.array(y)

print(x.shape,y.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x=x,y=y)
# plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
poly=PolynomialFeatures(degree=50)
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print("Training Data",X_train.shape,Y_train.shape)
print("Test Data",X_test.shape,Y_test.shape)
```

```python
X_train=X_train.reshape(-1,1)
X_test=X_test.reshape(-1,1)
print("Training Data",X_train.shape,Y_train.shape)
print("Test Data",X_test.shape,Y_test.shape)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
# X_train_poly=X_train_poly.reshape(1,-1)

X_train_poly
```

```python
Y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train_poly,Y_train)
```

```python
y_pred=lr.predict(X_test_poly)
accuracy=r2_score(Y_test,y_pred)
accuracy
```
