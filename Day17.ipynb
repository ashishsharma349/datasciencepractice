## Gradient Boosting

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.tree import DecisionTreeRegressor

x0=np.linspace(10,100)
x1=np.linspace(100,200)
output=np.random.rand(1,50)
```

```python
# df=pd.DataFrame([x0,x1],columns=['x0','x1'])
df = pd.DataFrame({'x0': x0, 'x1': x1,'out':output.ravel()})
df_copy=df
df[:10]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

def GDBoost(df, num, alpha):
    # Step 1: Initialize base prediction
    df["fx_not"] = df["out"].mean()
    df["pred"] = df["fx_not"]  # initial prediction (constant mean)
    
    # Step 2: Iteratively build trees
    for i in range(num):
        residual = df["out"] - df["pred"]
        X = df[["x0", "x1"]]
        y = residual
        dt = DecisionTreeRegressor(max_depth=8)
        dt.fit(X, y)
        
        y_pred = dt.predict(X)
        df[f"fx_{i+1}"] = y_pred
        df["pred"] = df["pred"] + alpha * y_pred

        df[f"ri_{i+1}"] = residual

    return df

df=GDBoost(df,10,0.1)
```
